```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x,y = make_blobs(n_samples=500, centers=4, cluster_std=0.5, random_state=0)
print(x.shape)
print(y.shape)
```

```python
from turtle import color


plt.plot(x[:,0], x[:,1], 'o')
plt.show()
```

```python
model = KMeans(n_clusters=4, random_state=0)
```

```python
model.fit(x)
center = model.cluster_centers_
print(center)
#[x_pos,y_pos]
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1], c=y_pred)
plt.scatter(center[0:,0], center[0:,1], c='yellow', label='Centroid 1')
plt.scatter(center[1:,0], center[1:,1], c='orange', label='Centroid 2')
plt.scatter(center[2:,0], center[2:,1], c='red', label='Centroid 3')
plt.scatter(center[3:,0], center[3:,1], c='black', label='Centroid 4')
plt.legend()
plt.show()
```
